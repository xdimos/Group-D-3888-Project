```python
import networkx as nx

import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

import networkx.algorithms.community as nx_comm
```

```python
G0 = nx.read_weighted_edgelist("./4932.protein.links.v11.5.txt",comments="#",nodetype=str)
```

```python
# delete those edges with a combined score of < threshold_score (small confidence)
threshold_score = 700
for edge in G0.edges: 
    weight = list(G0.get_edge_data(edge[0],edge[1]).values())
    if(weight[0] < threshold_score):
        G0.remove_edge(edge[0],edge[1])
        
# code taken from the ed workshop
```

```python
subcenters = nx.subgraph_centrality(G0)
```

```python
subcenters
```

```python
ls = list(subcenters.values())
```

```python
plt.hist(ls, bins = np.arange(0,1.4*1e89,0.1*1e89))
plt.title('All subgraph centralitys')
plt.show()
```

```python
arr = np.array(ls)
bigarr = arr[arr>0.1*1e89]
```

```python
plt.hist(bigarr)
plt.title('Larger subgraph centralities')
plt.show()
```

```python
smallarr = arr[arr<=0.1*1e89]
plt.hist(smallarr)
plt.title('Smaller subgraph centralities')
plt.show()
```
